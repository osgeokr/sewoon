```python
# 라이브러리 추가
import folium
```

```python
# 마커: 팝업, 툴팁 설정
m = folium.Map(
    location=[37.569556, 126.995229],
    zoom_start=20,
    tiles='http://api.vworld.kr/req/wmts/1.0.0/D05C77C9-AB62-3E70-9183-0E044A461BBD/Base/{z}/{y}/{x}.png',
    attr='VworldBase'
)

tooltip = 'Click me!'

folium.Marker([37.569556, 126.995229], popup='<b>Sewoon Sangga</b>', tooltip=tooltip).add_to(m)

m
```

```python
# 마커: 아이콘 설정
m = folium.Map(
    location=[37.569556, 126.995229],
    zoom_start=20,
    tiles='http://api.vworld.kr/req/wmts/1.0.0/D05C77C9-AB62-3E70-9183-0E044A461BBD/Base/{z}/{y}/{x}.png',
    attr='VworldBase'
)

folium.Marker(
    location=[37.569556, 126.995229],
    popup='Sewoon Sangga',
    icon=folium.Icon(icon='cloud')
).add_to(m)

folium.Marker(
    location=[37.570514, 126.995048],
    popup='Makercity Sewoon',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(m)

m
```

```python
# 마커: 반경 설정
m = folium.Map(
    location=[37.569556, 126.995229],
    zoom_start=20,
    tiles='http://api.vworld.kr/req/wmts/1.0.0/D05C77C9-AB62-3E70-9183-0E044A461BBD/gray/{z}/{y}/{x}.png',
    attr='VworldGray'
)

# Circle
folium.Circle(
    radius=50,
    location=[37.569556, 126.995229],
    popup='Sewoon Sangga',
    color='#3186cc',
    fill=True,
    fill_color='#3186cc'
).add_to(m)

#CircleMarker
folium.CircleMarker(
    location=[37.570514, 126.995048],
    radius=10,
    popup='Makercity Sewoon',
    color='crimson',
    fill=False,
).add_to(m)

m
```

```python
# 마커: 팝오버(popover)
m = folium.Map(
    location=[37.569556, 126.995229],
    zoom_start=20,
    tiles='http://api.vworld.kr/req/wmts/1.0.0/D05C77C9-AB62-3E70-9183-0E044A461BBD/Base/{z}/{y}/{x}.png',
    attr='VworldBase'
)

m.add_child(folium.LatLngPopup())

m
```
